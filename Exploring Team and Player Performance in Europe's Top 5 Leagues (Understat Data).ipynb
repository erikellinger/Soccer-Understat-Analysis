```python
# Load libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load team and player data
team_df = pd.read_csv("output/understat_team_stats_2023_2024.csv")
player_df = pd.read_csv("output/understat_player_stats_2023_2024.csv")

# ------------------- TEAM xG vs Goals -------------------

# Extract xG and goals from nested dicts in team_df
def extract_goal_value(row):
    try:
        return float(eval(row["goals"])[row["side"]])
    except:
        return None

def extract_xg_value(row):
    try:
        return float(eval(row["xG"])[row["side"]])
    except:
        return None

team_df["goals_numeric"] = team_df.apply(extract_goal_value, axis=1)
team_df["xG_numeric"] = team_df.apply(extract_xg_value, axis=1)

# Group and calculate overperformance
team_summary = team_df.groupby(["Team", "Season"]).agg({
    "xG_numeric": "sum",
    "goals_numeric": "sum"
}).reset_index()
team_summary["xG_diff"] = team_summary["goals_numeric"] - team_summary["xG_numeric"]

# Plot: Team Overperformance
plt.figure(figsize=(12, 6))
sns.barplot(data=team_summary.sort_values("xG_diff", ascending=False),
            x="xG_diff", y="Team", hue="Season", dodge=True)
plt.axvline(0, color="black", linestyle="--", linewidth=1)
plt.title("Team Overperformance (Goals - xG)")
plt.xlabel("Overperformance")
plt.ylabel("Team")
plt.tight_layout()
plt.show()

# ------------------- PLAYER xG Chart -------------------
# Plot: Top 15 Players by xG (Line + Dot, No Team Hue)
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=top_xg_players,
    x="xG",
    y="player_name",
    marker="o",
    linewidth=2,
    color="steelblue"
)
plt.title("Top 15 Players by Expected Goals (xG)")
plt.xlabel("Total xG")
plt.ylabel("Player")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Ensure xG and goals are numeric
player_df["xG"] = pd.to_numeric(player_df["xG"], errors="coerce")
player_df["goals"] = pd.to_numeric(player_df["goals"], errors="coerce")

# Group by player and aggregate both xG and goals
xg_goals_summary = (
    player_df.groupby("player_name")
    .agg({"xG": "sum", "goals": "sum"})
    .reset_index()
    .sort_values("xG", ascending=False)
    .head(15)
)

# Plot: xG vs Goals for top 15 players with annotations
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=xg_goals_summary,
    x="xG",
    y="goals",
    s=100,
    color="dodgerblue"
)

# Add annotation for each player
for _, row in xg_goals_summary.iterrows():
    plt.text(row["xG"] + 0.1, row["goals"], row["player_name"], fontsize=9)

# Add reference line (xG = Goals)
plt.plot([xg_goals_summary["xG"].min(), xg_goals_summary["xG"].max()],
         [xg_goals_summary["xG"].min(), xg_goals_summary["xG"].max()],
         'k--', label="Expected = Actual")

plt.title("Top 15 Players: Expected Goals vs Actual Goals")
plt.xlabel("Total xG")
plt.ylabel("Total Goals")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Aggregate xG and Goals by Player, Team, League
xg_goals_summary = player_df.groupby(["player_name", "Team", "League"]).agg({
    "xG": "sum",
    "goals": "sum"
}).reset_index()

# Convert to numeric
xg_goals_summary["xG"] = pd.to_numeric(xg_goals_summary["xG"], errors="coerce")
xg_goals_summary["goals"] = pd.to_numeric(xg_goals_summary["goals"], errors="coerce")

# Drop missing values
xg_goals_summary = xg_goals_summary.dropna(subset=["xG", "goals"])

# Keep top 10 players per league by xG
top10_per_league = xg_goals_summary.groupby("League", group_keys=False).apply(
    lambda x: x.sort_values("xG", ascending=False).head(10)
).reset_index(drop=True)

# Create FacetGrid: xG vs Goals, faceted by League
g = sns.FacetGrid(top10_per_league, col="League", col_wrap=3, height=4, sharex=False, sharey=False)

g.map_dataframe(sns.scatterplot, x="xG", y="goals", hue="player_name", s=70, legend=False)

# Annotate and plot identity line
for ax, (_, subdata) in zip(g.axes.flat, top10_per_league.groupby("League")):
    min_val = min(subdata["xG"].min(), subdata["goals"].min())
    max_val = max(subdata["xG"].max(), subdata["goals"].max())
    ax.plot([min_val, max_val], [min_val, max_val], 'k--', alpha=0.6)
    for _, row in subdata.iterrows():
        ax.text(row["xG"] + 0.1, row["goals"], row["player_name"], fontsize=8)

g.set_axis_labels("Expected Goals (xG)", "Goals")
g.set_titles("{col_name}")
plt.suptitle("Top 10 Player xG vs Goals by League", y=1.05, fontsize=16)
plt.tight_layout()
plt.show()
```
